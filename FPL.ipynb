# FPL Analysis

```python
import requests
import pandas as pd
import numpy as np
```

### Import Data with API

```python
url = 'https://fantasy.premierleague.com/api/bootstrap-static/'
req = requests.get(url)
json = req.json()
json.keys()
```

### Convert API Data to DataFrame 

```python
elements = pd.DataFrame(json['elements'])
elements_types = pd.DataFrame(json['element_types'])
teams = pd.DataFrame(json['teams'])
```

```python
elements.head()
```

```python
elements_types.head()
```

```python
teams.head()
```

#### Trimming Data 

```python
Trim_elements = elements[['second_name','team','element_type','selected_by_percent','now_cost','minutes','transfers_in','value_season','total_points']]

Trim_elements.head()
```

### Working with the Dataframe

```python
Trim_elements['position'] = Trim_elements['element_type'].map(elements_types.set_index('id').singular_name)

Trim_elements.head()
```

```python
Trim_elements['team'] = Trim_elements['team'].map(teams.set_index('id').name)

Trim_elements.head()
```

#### Type Casting Value and Sorting by Value

```python
Trim_elements['value'] = Trim_elements.value_season.astype(float)

Trim_elements.sort_values('value',ascending=False).head(10)
```

#### Sorting by Total Points

```python
Trim_elements.sort_values('total_points',ascending=False).head(10)
```

#### Removing 0 Time Players

```python
Trim_elements = Trim_elements.loc[Trim_elements.value > 0]
```

#### Segragating Based on position

```python
pivot = Trim_elements.pivot_table(index='position',values='value',aggfunc=np.mean).reset_index()

pivot.sort_values('value',ascending=False)
```

#### Which Team Providing Max Value

```python
best_team = Trim_elements.pivot_table(index='team', values='value', aggfunc=np.mean).reset_index()
best_team.sort_values('value', ascending=False)
```
